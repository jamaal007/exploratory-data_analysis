Normalization mean to bring the data to a range which make that easy to compare. 
   - Uniform the DATA
   - They have same impact
   - To reduced computational resources

The above data is really in a wide range and we need to normalize and hard to compare in this format. Normalization change the values range from 0 to 1. After normalization have similar influence in our models.
 - Methods for normalization
   - **Simple feature scalling**
     - X(new)=x(old)x(max)
   - **Min-Max method**
     - x-x(min)/x(max)-x(min)
   - **Z-Score method**
     - X=x-x.mean/x.std
   - **Log Transformation**
     - x=np.log[x]

## **Normalizing fare first**

**1 - Feature Scalling Method**     (0 to 1)

```python
# load fresh data set 
ks2=sns.load_dataset("titanic")
ks2.head()
```

```python
# 1-Sample feature scalling

# for "fare"

ks2["fare"]=ks2["fare"]/ks2["fare"].max()
ks2.head()
```

```python
# 1-Sample feature scalling

# for "age"

ks2["age"]=ks2["age"]/ks2["age"].max()
ks2.head()
```

If we see **fare** in both the tables and compared before normalization and after, it is being **Normalized**. with method No.1

**2- Min Max Method**    (0 to 1)

```python
# load fresh data set 
ks2=sns.load_dataset("titanic")
ks2.head()
```

```python
# 2- min max method 

# for "fare"

ks2["fare"]=(ks2["fare"]-ks2["fare"].min())/(ks2["fare"].max()-ks2["fare"].min())
ks2.head()
```

```python
# 2- min max method 

# for "age"

ks2["age"]=(ks2["age"]-ks2["age"].min())/(ks2["age"].max()-ks2["age"].min())
ks2.head()
```

Normalized with method No.2.

**3- Z Score Method**    (-3 to 3)

```python
# load the data set 
ks3=sns.load_dataset("titanic")

#  Z-score (standard test)    

# for "fare"

ks3["fare"]=(ks3["fare"]-ks3["fare"].mean())/ks3["fare"].std()
ks3.head()
```

```python
#  Z-score (standard test)    

# for "age"

ks3["age"]=(ks3["age"]-ks3["age"].mean())/ks3["age"].std()
ks3.head()
```

**4- Log Transformation Method** 

```python
# load fresh data set 
ks4=sns.load_dataset("titanic")
ks4.head()
```

```python
# 4- log transformation 

# for "fare"

ks4["fare"]=np.log(ks4["fare"])
ks4.head()
```

```python
# 4- log transformation 

# for "age"

ks4["age"]=np.log(ks4["age"])
ks4.head()
```
